```python
import pandas as pd
import altair as alt
import warnings


alt.data_transformers.disable_max_rows()
warnings.simplefilter(action='ignore', category=FutureWarning)

# Loading Data
path = "..\\data\\raw\\play_by_play_2022.csv"
df = pd.read_csv(path, low_memory=False)
```

```python
# Not all variables are listed by df.describe() 
#df.describe().T
```

```python
# Select the relevant variables, the lesser we select the more we have
data_selection = ["home_team", "away_team","game_seconds_remaining", "yardline_100", 
"down", "total_home_score", "score_differential", "play_clock", "touchdown", "yards_gained" ]

df_small = df[data_selection]


# drop missing values
df_small.dropna(inplace=True)#
# filter Data Only where yard are gained
df_small = df_small.loc[df.home_team=="NYJ" ]

df_small.loc[:10, data_selection]
```

```python
df_small.info()
#df_small = df_small['home_team'].astype("nominal")
```

```python
# Show data with a touchdown
df_small.loc[df.touchdown > 0].iloc[0:10]
```

```python
alt.Chart(df_small).mark_point().encode(
    x=alt.X("score_differential"),
    y=alt.Y('yards_gained'),
    #color=alt.Color('play_clock')
)
```

```python
alt.Chart(df_small).mark_point().encode(
    x=alt.X("yardline_100"),
    y=alt.Y('yards_gained'),
    #color=alt.Color('play_clock')
)
```

```python
alt.Chart(df_small).mark_boxplot().encode(
    y=alt.Y("home_team", scale=alt.Scale(type="symlog")),
    x=alt.X('yards_gained'),
    color=alt.Color('home_team')
).properties(
    width=400,
    height=500,
)
```
